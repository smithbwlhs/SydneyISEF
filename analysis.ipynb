```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('boxplot.csv')

df.head()
grouped_data = df.groupby('Treatment')
sns.boxplot(data = df, x = 'Treatment', y = 'AU',showfliers = False)
plt.xlabel(None)
plt.ylabel("Absorbance units")
plt.savefig("430nm_boxplot",dpi=600)
plt.show()
```
